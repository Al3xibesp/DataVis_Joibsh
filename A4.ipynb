```python
import pandas as pd
import numpy as np

path1 = '/content/drive/MyDrive/Visualización de Datos/datasets/antropometria-dataset-2.csv'
path2 = '/content/drive/MyDrive/Visualización de Datos/datasets/adult.csv'

df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)
df1.head()
```

```python
df2.head()
```

Generar Tabla Dinamica que muestre el peso promedio de personas por estado

```python
pd.pivot_table(df1,
               values = 'peso',
               index = 'desc_ent',
               aggfunc = 'mean')
```

```python
# Agregue columnas que indiquen el peso promedio por estado
EstadoSexo = pd.pivot_table(df1,
               values = 'peso',
               index = 'desc_ent',
               columns = 'sexo',
               aggfunc = ['mean', 'median'])
EstadoSexo
```

Podemos ver claramente como los hombres tienen un peso mayor que el de las muejeres en todos los estados

```python
# Peso median y avg de sex 1 de Yucatan
EstadoSexo.loc['YUCATAN', [('mean', 1), ('median', 1)]]
```

```python
max_peso_sexo2_por_estado = df1[df1['sexo'] == 2].groupby('desc_ent')['peso'].max()
display(max_peso_sexo2_por_estado)
```

```python


# Valor maximo de los promedios en la tabla dinamica EstadoSexo
max_of_means = EstadoSexo['mean'].max().max()
print(max_of_means)
```

```python
max_of_means_sexo2 = EstadoSexo['mean'][2].max()
print(max_of_means_sexo2)
```
